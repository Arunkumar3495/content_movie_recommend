# data preprocessing

```python
# convert lowercase and remove numbers,punctuations spaces, etc
```

```python
df['clean_plot'] = df['Plot'].str.lower()
```

```python
df['clean_plot'][0:11]
```

```python
df['clean_plot'] = df['clean_plot'].apply(lambda x:re.sub('[^a-zA-Z]',' ',x))
df['clean_plot'][0:11]
```

```python
df['clean_plot'] = df['clean_plot'].apply(lambda x:re.sub('\s+',' ',x))
df['clean_plot'][0:11]
```

```python
# tokenize the sentences
df['clean_plot'] = df['clean_plot'].apply(lambda x:nltk.word_tokenize(x))
df['clean_plot'][0:11]
```

```python
# remove stop words
stop_words = nltk.corpus.stopwords.words('english')
plot = []
for sentence in df['clean_plot']:
    temp=[]
    for word in sentence:
        if word not in stop_words or len(word) >=3:
            temp.append(word)
    plot.append(temp)
plot
```

```python
df['clean_plot'] = plot
df['clean_plot'][0:11]
```

```python
df['Genre'][0]
```

```python
df['Genre'] = df['Genre'].apply(lambda x:x.split(','))
```

```python
df['Actors'][0]
```

```python
#df['Actors'] = df['Actors'].apply(lambda x:x.split(',')[:4])
df['Director'] = df['Director'].apply(lambda x:x.split(','))
```

```python
df['Actors'][:4]
```

```python
df['Director'][0:4]
```

```python
def clean(sentence):
    temp=[]
    for word in sentence:
        temp.append(word.lower().replace(' ',''))
    return temp
```

```python
df['Genre'][:2]
```

```python
df['Genre'] =[clean(x) for x in df['Genre']]
```

```python
df['Genre'][:2]
```

```python
df['Actors'] =[clean(x) for x in df['Actors']]
df['Director'] =[clean(x) for x in df['Director']]
```

```python
df['Actors'][0]
```

```python
df['Director'][:3]
```

```python
# combining all the column
columns = ['clean_plot','Genre','Actors','Director']
l=[]
for i in range(len(df)):
    words = ''
    for col in columns:
        words += ' '.join(df[col][i])+' '
    l.append(words)
l
```

```python
df['clean_input'] = l
df1 = df[['Title','clean_input']]
df1.head()
```

```python
# feature extraction
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
tfidf = TfidfVectorizer()
features = tfidf.fit_transform(df1['clean_input'])
```

```python
# creat cosine similarity matrix
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(features,features)
print(cosine_sim)
```

```python
# movie recommendation
index = pd.Series(df1['Title'])
index
```

```python
def recommand_movies(title):
    movies= []
    idx = index[index == title].index[0]
    print(idx)
    score = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    top10 = list(score.iloc[1:11].index)
    print(top10)
    for i in top10:
        movies.append(df['Title'][i])
    return movies
```

```python
recommand_movies('Spider-Man 3')
```
